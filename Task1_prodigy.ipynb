```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('task1_csv.csv')
df
```

#**PLOTTING** **COUNT** **PLOT**

```python

#plotting count plot
sns.countplot(x="cp", data=df)
sns.countplot(x="cp",hue="sex",data=df)
```

```python
sns.countplot(x="age", data=df)
sns.countplot(x="age",hue="sex",data=df)
```

#**BOX** **PLOT**

```python
sns.boxplot(x="sex", y="chol",data=df)
```

```python
sns.boxplot(x="age", y="chol",data=df)
```

**Creating a scatter plot to visualize the relationship between two numerical variables**


```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['chol'], color='green', alpha=0.5)
plt.title('Age vs Chol')
plt.xlabel('age')
plt.ylabel('chol')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['sex'], df['cp'], color='green', alpha=0.5)
plt.title('sex vs cp')
plt.xlabel('sex')
plt.ylabel('cp')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

**Creating a pie chart for the distribution of a categorical variable**




```python
plt.figure(figsize=(8, 8))
df['target'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title(' Target Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['chol'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title(' cholestral Distribution')
plt.ylabel('')
plt.show()
```
